```python
def detect_ordinal(data,cat_cols,keywords=None,max_unique=7):
    if keywords is None:
        keywords={'Ex','Gd','TA','Fa','Po','None'}
    ordinal_cols=[]
    for col in cat_cols:
        unique_vals=set(data[col].dropna().unique())
        if len(unique_vals.intersection(keywords))>=2 :
            ordinal_cols.append(col)
    return ordinal_cols
ordinal_cols=detect_ordinal(data,cat_cols)
nominal_cols=[col for col in cat_cols if col not in ordinal_cols]
    

```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
rf=RandomForestRegressor()
param_grid={
    'n_estimators':[100,200],
    'max_depth':[None,10,20],
    'min_samples_split':[2,5]
}
grid_rf=GridSearchCV(rf,param_grid,cv=5,scoring='r2',n_jobs=-1)    
grid_rf.fit(X_train,y_train)
best_rf=grid_rf.best_estimator_
y_pred_rf=best_rf.predict(X_val)

from sklearn.model_selection import RandomizedSearchCV
from xgboost import XGBRegressor
xgb=XGBRegressor()
param_dist={
    'n_estimators':[100,200,300],
    'learning_rate':[0.01,0.05,0.1],
    'max_depth':[3,5,7],
    'subsample':[0.7,0.8,1.0]}
random_xgb=RandomizedSearchCV(xgb,param_distributions=param_dist,n_iter=10,cv=5,scoring='r2',n_jobs=-1)
    
    
random_xgb.fit(X_train,y_train)
best_xgb=random_xgb.best_estimator_
y_pred_xgb=best_xgb.predict(X_val)
    


#lightgbm
from lightgbm import LGBMRegressor
lgbm=LGBMRegressor()
param_dist={'n_estimators':[100,200],'learning_rate':[0.01,0.1],'num_leaves':[31,50,100]}
random_lgbm=RandomizedSearchCV(lgbm,param_distributions=param_dist,n_iter=10,cv=5,scoring='r2',n_jobs=-1)
random_lgbm.fit(X_train,y_train)
best_lgbm=random_lgbm.best_estimator_
y_pred_lgbm=best_lgbm.predict(X_val)
#catboost regressor
from catboost import CatBoostRegressor
cat=CatBoostRegressor(verbose=0)
param_grid={
    'iterations':[100,200],
    'depth':[4,6,8],
    'learning_rate':[0.01,0.1]
}
grid_cat=GridSearchCV(cat,param_grid,cv=5,scoring='r2',n_jobs=-1)
grid_cat.fit(X_train,y_train)
best_cat=grid_cat.best_estimator_
y_pred_cat=best_cat.predict(X_val)






```

```python
#imputing nan values in categorical features because below models cannot handle nan even in categorical features
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='most_frequent')
X_train=imputer.fit_transform(X_train)
X_val=imputer.transform(X_val)


```

```python
#since linear regression is a optimizatioon based technique,scaling improves performance

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_val=scaler.transform(X_val)
#training a KNN for regression
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor()
param_grid={
    'n_neighbors':[3,5,7],
    'weights':['uniform','distance']}
grid_knn=GridSearchCV(knn,param_grid,cv=5,scoring='r2')
grid_knn.fit(X_train,y_train)
best_knn=grid_knn.best_estimator_
y_pred_knn=best_knn.predict(X_val)
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
model=LinearRegression()
model.fit(X_train,y_train)
y_pred_linear=model.predict(X_val)
```

```python
ridge=Ridge()
param_grid_ridge={'alpha':[0.01,0.1,1,10]}
grid_ridge=GridSearchCV(ridge,param_grid_ridge,cv=5,scoring='r2')
grid_ridge.fit(X_train,y_train)
best_ridge=grid_ridge.best_estimator_
y_pred_ridge=best_ridge.predict(X_val)
```

```python
#lasso
param_grid_lasso={'alpha':[0.001,0.01,0.1,1]}
lasso=Lasso(max_iter=10000)
grid_lasso=GridSearchCV(lasso,param_grid_lasso,cv=5,scoring='r2')
grid_lasso.fit(X_train,y_train)
best_lasso=grid_lasso.best_estimator_
y_pred_lasso=best_lasso.predict(X_val)
```

```python
from sklearn.svm import SVR

from sklearn.preprocessing import StandardScaler
y_scaler=StandardScaler()
y_train_scaled=y_scaler.fit_transform(y_train.to_numpy().reshape(-1,1))#transform expects 2d matrix


#since the dataset is small and linear can apply svr
model1=SVR()
param_grid={
    'kernel':['linear','rbf','poly'],
    'C':[0.1,1,10],
    'epsilon':[0.01,0.1,0.5]
}
grid_svr=GridSearchCV(model1,param_grid,cv=5,scoring='r2')
grid_svr.fit(X_train,y_train_scaled.ravel())
best_svr=grid_svr.best_estimator_
y_pred_svr=best_svr.predict(X_val)

```

```python
y_pred_svr=y_scaler.inverse_transform(y_pred_svr.reshape(-1,1)).ravel()
```

```python
from sklearn.metrics import mean_squared_error,r2_score
import numpy as np
print("rmse" + str(np.sqrt(mean_squared_error(y_val,y_pred_rf))))
print("MSE:",mean_squared_error(y_val,y_pred_rf))
print("r2:",r2_score(y_val,y_pred_rf))

print("rmse", np.sqrt(mean_squared_error(y_val,y_pred_lgbm)))
print("MSE:",mean_squared_error(y_val,y_pred_lgbm))
print("r2:",r2_score(y_val,y_pred_lgbm))

print("rmse", np.sqrt(mean_squared_error(y_val,y_pred_xgb)))
print("MSE:",mean_squared_error(y_val,y_pred_xgb))
print("r2:",r2_score(y_val,y_pred_xgb))



print("rmse", np.sqrt(mean_squared_error(y_val,y_pred_cat)))
print("MSE:",mean_squared_error(y_val,y_pred_cat))
print("r2:",r2_score(y_val,y_pred_cat))


print("rmse", np.sqrt(mean_squared_error(y_val,y_pred_knn)))
print("MSE:",mean_squared_error(y_val,y_pred_knn))
print("r2:",r2_score(y_val,y_pred_knn))
                    



print("rmse", np.sqrt(mean_squared_error(y_val,y_pred_ridge)))
print("MSE:",mean_squared_error(y_val,y_pred_ridge))
print("r2:",r2_score(y_val,y_pred_ridge))




print("rmse", np.sqrt(mean_squared_error(y_val,y_pred_lasso)))
print("MSE:",mean_squared_error(y_val,y_pred_lasso))
print("r2:",r2_score(y_val,y_pred_lasso))


print("rmse", np.sqrt(mean_squared_error(y_val,y_pred_svr)))
print("MSE:",mean_squared_error(y_val,y_pred_svr))
print("r2:",r2_score(y_val,y_pred_svr))
```
